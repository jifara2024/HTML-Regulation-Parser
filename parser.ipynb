```python
import json
from bs4 import BeautifulSoup

with open("test.html", encoding="utf-8") as f:
    soup = BeautifulSoup(f, "html.parser")

title = soup.find("h1").text.strip()

chapter_data = []

for chapter in soup.find_all("h2"):
    chapter_item = {
        "chapter": chapter.text.strip(),
        "articles": []
    }

    node = chapter.find_next_sibling()

    while node and node.name != "h2":
        if node.name == "h3":
            article_title = node.text.strip()
            article_text = ""

            next_p = node.find_next_sibling("p")
            if next_p:
                article_text = next_p.text.strip()

            chapter_item["articles"].append({
                "article": article_title,
                "text": article_text
            })

        node = node.find_next_sibling()

    chapter_data.append(chapter_item)

result = {
    "title": title,
    "chapters": chapter_data
}

print(json.dumps(result, indent=4))
```

```python
with open("output.json", "w", encoding="utf-8") as f:
    json.dump(result, f, indent=4)

print("output.json saved")
```
